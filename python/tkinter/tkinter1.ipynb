```python
from  tkinter import *
import tkinter.messagebox
main_interface = Tk()

li     = ['C','python','php','html','SQL','java']
movie  = ['CSS','jQuery','Bootstrap']  #movie 电影
listb = Listbox(main_interface)
listb2 = Listbox(main_interface)
for item in li:
    listb.insert(0,item)

for item in movie:
    listb2.insert(0,item)
    
listb.pack()
listb2.pack()
main_interface.mainloop()
```
