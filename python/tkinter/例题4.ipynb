```python
# 
from tkinter import *
root = Tk()
root.title('')
Label(root,text='').grid(row=0,column=0) #row,column 
Label(root,text='').grid(row=1,column=0)

v1 = StringVar()
v2 = StringVar()

e1 =Entry(root,textvariable=v1)
e2 =Entry(root,textvariable=v2,show="*") #show*
e1.grid(row=0,column=1,padx=10,pady=5)
e2.grid(row=1,column=1,padx=10,pady=5)

def show():
    print("{}".format(e1.get()))
    print("{}".format(e2.get()))
Button(root,text="",command=show).grid(row=3, column=0, padx=10, pady=5)
Button(root, text='', width=10, command=root.quit) \
.grid(row=3, column=1, sticky=E, padx=10, pady=5)
mainloop()
```

```python
# 
from tkinter import *
root= Tk()
root.title("")
def test():
    if e1.get()=='':
        print("")
        e2.insert(0,"")
        return True
    else:
        print("")
        print(e1.get())
        e1.delete(0,END)
        return False
    
v = StringVar()
#focusouttest<tab>
e1 = Entry(root,textvariable=v,validate='focusout',validatecommand=test)
e2 =Entry(root)
e1.pack(padx=10,pady=10)
e2.pack(padx=10,pady=10)
mainloop()
```

```python
# 
#w
from tkinter import *
root = Tk()
root.title("")

frame =Frame(root)
frame.pack(padx=10,pady=10)
v1 = StringVar()
v2 = StringVar()
v3 = StringVar()

def test(content):
    return content.isdigit()
testCMD = frame.register(test)

e1 = Entry(frame, width=10, textvariable=v1, validate='key', \
validatecommand=(testCMD, '%P')).grid(row=0, column=0)  #width
Label(frame, text='+').grid(row=0, column=1)
e2 = Entry(frame, width=10, textvariable=v2, validate='key', \
           validatecommand=(testCMD, '%P')).grid(row=0, column=2)
Label(frame, text='=').grid(row=0, column=3)
e3 = Entry(frame, width=10, textvariable=v3,state="readonly").grid(row=0, column=4)
def calc():
    result = int(v1.get()) + int(v2.get())
    v3.set(str(result))
Button(frame, text='', command=calc).grid(row=1, column=2, pady=5)
mainloop()
```
