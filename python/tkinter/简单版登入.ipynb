```python
# 导入库
import os
from tkinter import *
import tkinter.messagebox
```

```python
#设置居中显示
def center_window(main, width, height):
    screenwidth = main.winfo_screenwidth()
    screenheight = main.winfo_screenheight()
    size = '%dx%d+%d+%d' % (width, height, (screenwidth - width) / 2, (screenheight - height) / 2)
    main.geometry(size)
    main.update()
```

```python
#切换用户界面函数
def quit(e):
    quit = tkinter.messagebox.askokcancel('提示', '是否要创建账号？')
    if quit == True:
        main.destroy()
        create()

def quit2(e):
    cancel = tkinter.messagebox.askokcancel('提示', '是否取消创建账号')
    if cancel == True:
        root.destroy()
        zhu()
```

```python
#登入界面
def zhu():
    global main
    main = Tk()
    #调用界面剧中显示
    center_window(main, 500, 340)
    #设置界面不能变化
    main.resizable(0, 0)
    #背景图片
    photo = PhotoImage(file ="pyimage2.gif")
    bg  = Label(main, image=photo,compound = CENTER)
    bg.pack()
    #标题
    main.title("登入")
    #图标

    main.iconbitmap('qqico.ico')
    #框架
    frame = Frame(main)
    frame.pack(side=LEFT)
    #账号，密码
    lb1 = Label(frame,text="账号",)
    lb2 = Label(frame,text="密码",)
    lb1.grid(row=0,column=0)
    lb2.grid(row=1,column=0)
    #文本部分
    e1 = Entry(frame)
    e1.grid(row=0,column=1)
    e2 = Entry(frame,show="*")
    e2.grid(row=1,column=1)
    #创建账户
    lb3 = Label(frame,text="创建账号",bg="red",width=5)
    lb3.grid(row=0,column=3)
    lb3.bind('<Button-1>',quit)#单击鼠标左键，绑定Jump()函数

    bt = Button(frame,text='登入',command = lambda :login(e1.get(),e2.get()))
    bt.grid(row=2,column=2)

    mainloop()
zhu()
```

```python
#创建用户界面
def create():    
    global root
    root = Tk()
    root.title("创建用户")
    root.iconbitmap('qqico.ico')
    center_window(root, 250,70)
    root.resizable(0, 0)
    frame = Frame(root)
    frame.pack(side=LEFT)
    lb1 = Label(frame,text="创建的账号",)
    lb2 = Label(frame,text="账号的密码",)
    lb1.grid(row=0,column=0,sticky=E+W)
    lb2.grid(row=1,column=0)
    e1 = Entry(frame)
    e1.grid(row=0,column=1)
    e2 = Entry(frame,show="*")
    e2.grid(row=1,column=1)
    bt1 = Button(frame,text='创建')
    bt1.grid(row=2,column=0,sticky=E)
    bt2 = Button(frame,text='返回')
    bt2.grid(row=2,column=1,sticky=E)    
    bt2.bind("<Button-1>",quit2)
    
    root.mainloop()
```
