```python
import requests
import parsel
from tqdm import tqdm
import pandas as pd
import  PySimpleGUI as sg
import time
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.64"}
def gui():
    novel_name = sg.popup_get_text('请输入你要下载的小说名字或者作者名字：')
    save_path = sg.popup_get_file('请输入保存的位置')
    main(novel_name,save_path)

def get_response(url): 
    res =requests.get(url=url,headers=headers)
    res.encoding=res.apparent_encoding
    return res

def save(name,title,content,novel_path):
    os.chdir(novel_path)
    with open(name+".txt","a",encoding="utf-8") as fp:
        fp.write(title)
        fp.write("\n")
        fp.write(content)
        fp.write("\n")

def get_all_novel(url,novel_path):
    response =get_response(url)
    selector =parsel.Selector(response.text)
    novel_name=selector.css("#info h1::text").get()
    href=selector.css('#list dd a::attr(href)').getall()
    for link in tqdm(href) :
        link_url="https://www.biquges.com/"+link     
        html_date=get_response(link_url).text
        selector_1=parsel.Selector(html_date)
        title=str(selector_1.css(".bookname h1::text").get())
        content_list= selector_1.css("#content::text").getall()
        content="".join(content_list)
        save(novel_name,title,content,novel_path)

def main(novel_name,save_path):
    while True:
        url="https://www.biquges.com/modules/article/search.php"
        data={
            "searchkey":novel_name,
            "searchtype":"articlename",
        }
        response_2=requests.post(url=url,data=data,headers=headers)
        response_2.encoding=response_2.apparent_encoding
        selector_2=parsel.Selector(response_2.text) 
        nr=selector_2.css("#nr")
        
        if nr:
            lis=[]
            for index in nr:
                name=index.css("#nr .odd a::text").get()
                novel_id=index.css("#nr .odd a::attr(href)").get().split('/')[1]
                author=index.css("#nr td:nth-child(3)::text").get()

                dit={
                    "书名":name,
                    "ID":novel_id,
                    "作者":author,
                }
                lis.append(dit)
            print(f"搜索到{len(lis)}条内容,结果如下")
            pd_list=pd.DataFrame(lis)
            print(pd_list)      
            print("请输入你要下载的小说序号:")
  
            num=input(pd_list)
            num_id=lis[int(num)]["ID"]
            novel_url=f'https://www.biquges.com/{num_id}/'
            
            get_all_novel(novel_url,save_path)
            print("已经下载完毕")
            again=input("是否继续下载小说(y/n):")
            if again=="y":
                continue
            else:
                print("感谢您的使用")
                break
        else:
            print("抱歉,搜索没有结果,请重新输入")
            break
    
if __name__ == "__main__": 
    gui()
    
```
