```python
#这里用多进程的时候，思考是用页数来当一个线程，还是一个图片
import requests
from lxml import etree
import os
import time
import threading

headers = {
"User-Agent":
" Mozilla/5.0 (Windows NT 10.0;Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36 Edg/92.0.902.73"}
#有次调试因为这个没有当全局出错！！！

def save_address() :
    a =  os.getcwd().split("\\")[-1]
    if a != "图片" :
        if os.path.exists("图潘") :
            os.mkdir("图片")
            os.chdir("图片")
        else :
            os.chdir("图片")
        print(os.getcwd)



#print(len(picture_url))

def main():
    
    save_address()
    #url = "https://www.fabiaoqing.com/biaoqing"
  
    for i in range(1,3):        
        url = "https://www.fabiaoqing.com/biaoqing/lists/page/{}.html".format(i)        
        res = requests.get(url,headers)
        
        b = threading.Thread(target=parsing,args=(res,))   #主要的两条代码
        b.start()
        #parsing(res)  #不用多线程的代码
        
        
def parsing(res):
    
    html = etree.HTML(res.text)
    picture_url = html.xpath('//div[@class="tagbqppdiv"]/a/img')    
    for i in picture_url :
        url = i.xpath('./@data-original')[0]
        name = i.xpath(r'./@alt')[0]
        
        print(url,name)
        
        try:
            save(url,name)
        except:
            print(name+"  下载失败！！！") #下载失败的原因是名字里有空格什么的，可以用spilt去除等等
            continue
    print("over")

def save(url,name):    
    res_picture = requests.get(url,headers).content
    with open(name+".jpg",'wb') as f:
        f.write(res_picture)
    print(name+"  下载完成")
    
if __name__ == "__main__" :
    
    main()
    
    
    
```
